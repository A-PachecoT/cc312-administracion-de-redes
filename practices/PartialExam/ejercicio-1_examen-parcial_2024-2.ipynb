# Introducción

Las API (Application Program Interface) se utilizan generalmente para enviar y recibir datos de servidores web remotos. Para usar un API, se realiza una llamada a un servidor web y se recibe en la respuesta los datos que se solicitan.

Las APIs se alojan en servidores web. Cuando escribimos www.uni.edu.pe en la barra de direcciones de nuestro navegador, la  computadora solicita a www.uni.edu.pe accesar a una página web, esta es devuelta a nuestro navegador.

Las APIs funcionan de la misma forma, sólo que en lugar de solicitar al navegador web una página web, solicitamos datos. Los datos son devueltos generalmente en formato JSON.

## Sugerencia


Para integrar el uso de apis en nuestros programas utilizaremos la librería de python `requests`

- <i class="fa fa-link" aria-hidden="true"></i> https://www.w3schools.com/python/module_requests.asp

## Tipos de peticiones

Hay distintos tipos de peticiones, aunque la más común es la de tipo GET, que es la que utilizaremos para solicitar datos.

Un **endpoint** es una ruta en el servidor que se puede usar para obtener datos desde la API. Por ejemplo, `/users`  podría ser un endpoint de una API que nos daría información acerca de los usuarios. Para acceder a los endpoints, debemos añadir a la **url base** de la API el endpoint al que queremos acceder.

## Códigos de respuesta

Las peticiones a un servidor web devolverán un código de respuesta. Los códigos de respuesta indican qué ha pasado con la petición. A continuación se muestran algunos de los códigos de respuesta más comunes:

* 200: Todo ha ido bien y el resultado (si lo hay) ha sido devuelvo
* 301: El servidor te está dirigiendo a un endpoint diferente. Esto puede ocurrir cuando el API ha cambiado de dominio o de nombre.
* 400: El servidor devuelve error porque el formato de la petición no es correcto
* 401: El servidor devuelve error porque no estás correctamente autentificado.
* 403: El servidor devuelve error porque no tienes permisos para acceder al recurso que estás solicitando.
* 404: El servidor devuelve error porque el recurso solicitado no se encuentra en el servidor
* 500: Error interno del servidor

Ahora vamos a buscar información sobre el Perú.

````python
res = requests.get("https://covid-api.com/api/regions")

covid_regions = json.loads(res.text)
covid_regions
````

````python
for region in covid_regions['data']:
    if region['name'] == 'Peru':
        print(region)
````

Listo! Ahora sabemos que el iso de Perú es **PER**
Ahora vamos a buscar el resumen de Perú.

El endpoint para obtener el resumen de Perú es:

```
https://covid-api.com/api/reports?date=&q=&iso=&region_name=&region_province=
```
Para nuestro caso, la consulta sería:

```
https://covid-api.com/api/reports?date=2021-04-14&iso=PER
```
Porque no queremos filtrar por región ni por provincia.


````python
res = requests.get("https://covid-api.com/api/reports?date=2021-04-14&iso=PER")
print("Status code:", res.status_code)
print("Response:", res.text)
````

Habían algunas fechas sin datos, así que probé con fechas hasta que encontré una con datos en 2021-04-14.

````python
covid_peru = res.json()
covid_peru
````

Usando una lista de comprensión obtenga el resumen del Perú. Sugerencia:[**expresion** for **item** in **iterable** if **condicion**]

````python
resumen_peru_provincias = [{
    'confirmed': item['confirmed'],
    'deaths': item['deaths'],
    'recovered': item['recovered']
} for item in covid_peru['data']]
resumen_peru_provincias[:5]
````

````python
# Inicializar las sumas
total_confirmed = 0
total_deaths = 0
total_recovered = 0

# Sumar los valores
for item in resumen_peru_provincias:
    total_confirmed += int(item['confirmed'])
    total_deaths += int(item['deaths'])
    total_recovered += int(item['recovered'])

resumen_peru = {
    'country': 'Peru',
    'CountryCode': 'PE',
    'confirmed': total_confirmed,
    'deaths': total_deaths,
    'recovered': total_recovered
}
resumen_peru
````

## Parámetros de ruta

La información anterior era demasiado compleja y extensa, nosotros queremos acceder de manera estructurada, ya que solo queremos consultar nuestro país.

Por suerte la API dispone de endpoints específicos para los recursos que necesitamos, para ello usaremos parametros de ruta.

En una API de tipo REST identificamos los distintos recursos a los que vamos a acceder utilizando la ruta de la llamada. Generalmente hay partes de la llamada que son estáticos a los que denominamos **recursos** y partes que son dinámicas a las que denominamos parámetros de ruta o *path parameters*.

Por ejemplo:

- `GET /country`
- `GET /country/peru`

````python
# Felizmente para esta sección del laboratorio yo ya encontré el endpoint
# que me da los países.

res = requests.get("https://covid-api.com/api/regions")

covid_regions = json.loads(res.text)
covid_regions['data']
````

## Parámetros de búsqueda

La API anterior sigue suministrando demasiados datos, ya que nos muestra fechas desde el inicio de año y nosotros necesitamos solo unos meses concretos.

En algunas ocasiones se pueden añadir parámetros de consulta (**query parameters**) a las llamadas. Estos parámetros se añaden al final de la URL con el siguiente formato:

```
http://base_url/endpoint?param1=XXX&param2=YYY&...&paramN=ZZZ
```

Suelen utilizarse para filtrar o configurar las llamadas. Por ejemplo:

- GET /users?starts_with=ja&active=True

````python
# De igual manera, previamente hice consultas de parámetros,
# donde especifiqué el país Perú con el código PER
res = requests.get("https://covid-api.com/api/reports?date=2021-04-14&iso=PER")
print("Status code:", res.status_code)
print("Response:", res.text)
````

Muestre el documento JSON con los tres primeros elementos del diccionario referente al Perú

````python
res.json()['data'][:3]
````

Muestre el resumen del mes de mayo del 2022

````python
# Lastimosamente, la API no proporciona el parámetro de mes, solo de fecha.
# Así que tuve que hacer un for para iterar sobre el mes de mayo.
summary_mayo_2022 = []
for i in range(1, 32):
    res = requests.get(f"https://covid-api.com/api/reports?date=2022-05-{i:02}&iso=PER")
    if res.status_code == 200:
        summary_mayo_2022.append(res.json()['data'])
summary_mayo_2022[:3]
````

````python
# Inicializar las sumas
total_confirmed = 0
total_deaths = 0
total_recovered = 0

# Sumar los valores
for item in summary_mayo_2022[0]:
    total_confirmed += item['confirmed']
    total_deaths += item['deaths']
    total_recovered += item['recovered']

summary_mayo_2022 = {
    'country': 'Peru',
    'CountryCode': 'PE',
    'date': '2022-05',
    'confirmed': total_confirmed,
    'deaths': total_deaths,
    'recovered': total_recovered
}
summary_mayo_2022
````
